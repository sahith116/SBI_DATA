```python
#adding a separte column of datatime
data['datetime'] = pd.to_datetime(data['Date'] + ' ' + data['Time'], format='%d-%m-%Y %H:%M:%S') 
```

```python
data
```

```python
data= data.sort_values(by=['datetime'])
```

```python
#function that can calculate rank by making comparisions for the previous 5 days volume for current day
def calculate_rank(row):
    current_time = row['datetime']
    five_days_ago = current_time - pd.DateOffset(days=5)
    relevant_data = data[(data['datetime'] >= five_days_ago) & (data['datetime'] < current_time)]
    rank = (relevant_data['Volume'] > row['Volume']).sum() + 1
    return rank
```

```python
#more about calculate_rank function 
#->It takes a row of the dataframe as input.
#->It extracts the current datetime value from the input row.
#->It calculates the datetime value corresponding to five days ago from the current datetime.
#->It filters the dataframe to get the rows from the last 5 days (excluding the current day) with datetime values less than the current datetime.
#->It compares the volume of the input row with the volumes of the filtered rows.
#->It counts the number of volumes from the filtered rows that are greater than the volume of the input row. This count represents the rank of the volume for the input row.
#->It adds 1 to the count to get the actual rank (since rank starts from 1 for the highest volume).
```

```python
data['rank'] = data.apply(calculate_rank, axis=1)
```
